```python
# Implementing stack using object-oriented programming

class Stack:
    # Default constructor
    def __init__ (self):
        self.array = []
        self.length = 0

    # Push function that pushes a certain value passed as parameter
    def push (self, item):
        self.array.append(item)
        self.length += 1

    # Pop function that returns popped element and also handles error
    def pop (self):
        if self.is_empty():
            print ("Tried to pop an empty stack!")
            return -1
        else:
            top = self.array[-1]
            newarray = self.array[:-1]
            self.array = newarray
            self.length -= 1
            return top

    # Is_Empty function that returns if the stack is empty or not
    def is_empty (self):
        if (self.length == 0):
            return True
        else:
            return False

    def printstack (self):
        print (self.array)

# Driver to simulate some of these operations

s = Stack()

while (True):
    print ("The current stack is:", end =" ")
    s.printstack()
    print("What operations do you want to execute on the stack? Write 'Pop', 'Push', 'IsEmpty' or 'Exit':")
    query = input()

    if query.lower().replace(" ", "") == "pop":
        x = s.pop()
        if x != -1:
            print(f"The popped element is: {x}")
    elif query.lower().replace(" ", "") == "push":
        print("What do you want to push?")
        x = input()
        try:
            x = int(x)
        except:
            pass
        s.push(x)
    elif query.lower().replace(" ", "") == "isempty":
        if s.is_empty():
            print("Yes, the stack is empty.")
        else:
            print("No, the stack is not empty.")
    elif query.lower().replace(" ", "") == "exit":
        break
    else:
        print("Invalid input!")
```

```python
def reverse_string (s):
    st = Stack()
    for x in s:
        st.push(x)
    rev = ""
    while not st.is_empty():
        rev += st.pop()
    return rev

# Driver to execute this function
print("Enter a string that you want to reverse:")
str = input()
print("The reversed string is:")
print(reverse_string(str))
```
